#🛠️ Preparación de los Datos

##Importación de datos
Teniendo como base los datos tratados del el Challenge Telecom X parte 1, se genero el siguiente link desde el repositorio de GitHub de dicho challenge, produciendo un archivo CSV con el DF de nombre datos_tratados

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Luis-Rivadeneyra-94/TelecomX/refs/heads/main/datos_tratados.csv"
datos = pd.read_csv(url)
datos.head()
```

```python
datos.columns
```

```python
datos.info()
```

##Eliminación de Columnas Irrelevantes

```python
datos = datos.drop(columns=["Id Cliente"], axis=1)
```

```python
datos.head()
```

```python
datos.isnull().sum()
```

###Se encontro valores nulos en el DF en la columna "Permanencia de cliente" y "Total por cliente"

```python
datos = datos.dropna(subset=["Permanencia del cliente"])
datos = datos.dropna(subset=["Total por cliente"])
```

```python
datos.isnull().sum()
```

##Encoding

```python
datos = pd.get_dummies(datos, drop_first=True)
datos.info()
```
